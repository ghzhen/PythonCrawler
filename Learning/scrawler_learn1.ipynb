```python
import urllib2
import re

import urllib

import requests
```

```python
#爬虫闯关第一关

startUrl = "http://www.heibanke.com/lesson/crawler_ex00/"
response = urllib2.urlopen(startUrl)
spage = response.read()
match = re.search('你需要在网址后输入数字(\d+)', spage)
while match:
    dgts = match.group(1)
    print dgts
    response = urllib2.urlopen(startUrl + dgts)
    spage = response.read()
    match = re.search('<h3>下一个你需要输入的数字是(\d+)\..*</h3>', spage)
print spage
```

```python
#爬虫闯关第一关
import requests
import re

startUrl = "http://www.heibanke.com/lesson/crawler_ex00/"
r = requests.get(startUrl)
spage = r.text.encode('ascii','ignore')
match = re.search('<h3>.*(\d\d\d\d\d).*</h3>', spage)
while match:
    dgts = match.group(1)
    print dgts
    r = requests.get(startUrl + dgts)
    spage = r.text.encode('ascii','ignore')
    match = re.search('<h3>.*(\d\d\d\d\d).*</h3>', spage)
print spage
```

```python
# 爬虫闯关第2关
values = {}
url = 'http://www.heibanke.com/lesson/crawler_ex01/'
guess = 0
match = True
while(match and guess<=30):
    print guess
    values['username'] = 'z'
    values['password'] = str(guess)
    data = urllib.urlencode(values)
    request = urllib2.Request(url,data)
    response = urllib2.urlopen(request)
    text = response.read()
    guess = guess + 1
    match = re.search('<h3>您输入的密码错误, 请重新输入</h3>', text)
print text
```

```python
# 第二关 using ‘requests’ 
values = {}
url = 'http://www.heibanke.com/lesson/crawler_ex01/'
guess = 0
match = True
while(match and guess<=30):
    print guess
    values['username'] = 'z'
    values['password'] = str(guess)
    result = requests.post(url, data = values)
    text = result.text.encode('utf-8')
    guess = guess + 1
    match = re.search('<h3>您输入的密码错误, 请重新输入</h3>', text)
print text
```
