### 静雅思听首页 Homepage

```python
startUrl = "https://www.justing.com.cn"
response = urllib2.urlopen(startUrl)
spage = response.read()
print spage
```

### 排行榜 Ranklist

```python
# <a class="ranklist-link" data-for="ranklist" href="#/home/ranklist/books/popular">排行榜</a>
mstring = r"<a\sclass=\"ranklist-link\"\sdata-for=\"ranklist\"\shref=\"(.+)\">排行榜</a>"
match = re.search(mstring, spage)
ranklistUrl = startUrl + match.group(1)
response = urllib2.urlopen(startUrl)
rpage = response.read()
print rpage
```

### 图书类畅销榜 Bestseller

```python
# \u56fe\u4e66\n",
# href=\"/home/ranklist/books/bestseller\">\n",
# <span class=\"name\">\u7545\u9500\u699c</span>\n",
bookbestsellerUrl = startUrl + "/home/ranklist/books/bestseller"

response = urllib2.urlopen(bookbestsellerUrl)
state = 0
bestseller = DataFrame(columns=["Book Name","Link"])

for line in response:
    if state == 0:
        mstring = "<ul\sclass=\"list\slist-col2 main-ranklist-books-list-col2\">"
        match = re.search(mstring, line)
        if match:
            state = 1
            continue
    if state == 1:
        mstring = "</ul>"
        match = re.search(mstring, line)
        if match: 
            break;
        mstring = "<li\sclass=\"item\">"
        match = re.search(mstring, line)
        if match:
            state =2
            continue
    if state == 2:
        mstring = "<div\sclass=\"text\sname\">"
        match = re.search(mstring, line)
        if match:
            state = 3
            continue
    if state == 3:
        match = re.search("<a\shref=\"(.+)\">", line)
        if match:
            bookUrl = startUrl + match.group(1)
        else:
            match = re.search("^(?!</a>$).*$",line)
            if match:
                bookName = match.group(0)
                df = DataFrame([(bookName, bookUrl),], columns=["Book Name","Link"])
                bestseller = bestseller.append(df, ignore_index=True)
                state = 1
bestseller
```
